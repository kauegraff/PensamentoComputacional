###1 - Escreva um programa que lê um número inteiro N do usuário. O programa deverá escrever na tela o fatorial desse número (N!).
* Exemplo de entrada: 5
* Saída: 120
* Exemplo de entrada: 8
* Saída: 40320

```python
# Algoritmo que calcula o fatorial de um número

# Variavel
fatorial = 1

# Entrada dos dados
num = int(input("Digite um número inteiro: "))

# Loop do 1 até o número de entrada
for x in range(1, num+1):
  # Calculo e resultado do fatorial
  fatorial *= x
  print(fatorial)
```

### Sistema de segurança

```python
# Algoritmo que verifica a senha e libera acesso
import time

# Variaveis
senha = "4444"
senha_digitada = ""
cnt_tentativas = 0

# Executa o loop enquanto a senha inserida for diferente da senha
while senha_digitada != senha:
  # Entrada da senha
  senha_digitada = input("Digite a senha: ")
  cnt_tentativas += 1 # Contador de tentativas
  if senha_digitada == senha:
    break
  else:
    print("Senha inválida!")
  # Verifica se as tentativas são maior que três, se sim esperar determinado tempo para tentar novamente
  if cnt_tentativas >= 3:
    tempo = cnt_tentativas * 2
    print(f"3 tentatvas erradas! Aguarde {tempo}s para tentar novamente.")
    time.sleep(tempo) # 'dorme' por 60 segundos

print("Porta Liberada!")
```
